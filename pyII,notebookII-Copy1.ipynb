```python
def age_group(age):
    
    if age < 20:
        bucket = '<20'
    
    if age in range(20, 30):
        bucket = '20-29'
        
    if age in range(30, 40):
        bucket = '30-39'
        
    if age in range(40, 50):
        bucket = '40-49'
        
    if age in range(50, 60):
        bucket = '50-59'
   
    if age >= 60:
        bucket = '60+'

    return bucket
```

```python
df4['age_group'] = df4['Age'].apply(age_group)
```
